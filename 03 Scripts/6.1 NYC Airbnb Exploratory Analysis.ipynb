## 4. Data Wrangling

```python
#drop columns for host name due to personal identifiable information (PPI)
df.drop(columns = {'host name'}, inplace = True)
```

```python
#checking output
df.head()
```

```python
#correcting column names to be consistent 
df.columns=[col.lower().replace(" ","_") for col in df.columns]
df.columns
```

```python
#renaming columns 
df.rename(columns = {'name' : 'listing_title'}, inplace = True)
df.rename(columns = {'neighbourhood_group' : 'borough'}, inplace = True) 
df.rename(columns = {'neighbourhood' : 'neighborhood'}, inplace = True) 
df.rename(columns = {'lat' : 'latitude'}, inplace = True) 
df.rename(columns = {'long' : 'longitude'}, inplace = True) 
df.rename(columns = {'calculated_host_listings_count' : 'listings_per_host'}, inplace = True) 
```

```python
#replacing brookln with Brooklyn
df['borough'] = df['borough'].replace('brookln', 'Brooklyn')
df['borough'].value_counts()
```

```python
#replacing manhatan with Manhattan
df['borough'] = df['borough'].replace('manhatan', 'Manhattan')
df['borough'].value_counts()
```

### 5c. Changing Data Types

```python
#changing data type to datetime
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
#changing price and service fee
df['price']=df['price'].apply(lambda x: x.replace('$','') if pd.isna(x) == False else x)
df['service_fee']=df['service_fee'].apply(lambda x: x.replace('$','') if pd.isna(x) == False else x)
df['service_fee']=df['service_fee'].astype(float)
df['price']=df['price'].apply(lambda x: x.replace(',','') if pd.isna(x) == False else x)
df['price']=df['price'].astype(float)
```

```python
#changing data types
df['neighborhood'] = df['neighborhood'].astype('str')
df['borough'] = df['borough'].astype('str')
df['cancellation_policy'] = df['cancellation_policy'].astype('str')
df["number_of_reviews"] = np.array(df["number_of_reviews"], dtype = np.int16)
df["review_rate_number"] = np.array(df["review_rate_number"], dtype=np.int8)
df["listings_per_host"] = np.array(df["listings_per_host"], dtype = np.int16)
```
